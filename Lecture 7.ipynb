```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data 

x= [1,2,5,6,8,9]
y=[10,20,56,89,23,51]
```

```python
print(x,y)
```

```python
plt.bar(x,y)

plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.title("Simple Bar Plot")
plt.show()
```

```python
plt.plot(x,y)

plt.xlabel("x-axis label")
plt.ylabel("y-axis label")
plt.title("Simple Line Plot")
plt.show()
```
