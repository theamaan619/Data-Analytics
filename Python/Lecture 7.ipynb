```python
import pandas as pd
import openpyxl 
```

```python
data = pd.read_excel("empolyee Performance  Data.xlsx")
```

```python
data.head()
```

```python
data.info()
```

```python
mean_value = data[["Q1 Performance (%)","Q2 Performance (%)","Q3 Performance (%)","Q4 Performance (%)"]].mean()
print(mean_value)
```

```python
median_value = data[["Q1 Performance (%)","Q2 Performance (%)","Q3 Performance (%)","Q4 Performance (%)"]].median()
print(median_value)
```

```python
mode_value = data[["Q1 Performance (%)","Q2 Performance (%)","Q3 Performance (%)","Q4 Performance (%)"]].mode()
print(mode_value)
```

```python
quartely_avg=data.mean(numeric_only=True)

plt.figure(figsize=(8,5))
sns.barplot(x=quartely_avg.index, y=quartely_avg.values,palette="viridis")
plt.title("Average Performance Acroos Quater")
plt.ylabel("Performance (%)")
plt.xlabel("Quarters")
plt.show()
```

```python
df_melted=data.melt(id_vars=["Employee Name"],var_name="Quarter",value_name="Performance")
plt.figure(figsize=(10,6))
sns.lineplot(data=df_melted,x="Quarter",y="Performance",hue="Employee Name",marker="o")
plt.title("Performance Trends of Employee Across Quarters")
plt.ylabel("Performance (%)")
plt.xlabel("Quarters")
plt.legend(title="Employee Name")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df_melted,x="Quarter",y="Performance",palette="Set2")
plt.title("Performance Distribution Across Quarters")
plt.ylabel("Performance (%)")
plt.xlabel("Quarters")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(data.set_index("Employee Name").T,annot=True,cmap="coolwarm",fmt=".0f",cbar=True)
plt.title("Employee Performance Heatmap")
plt.ylabel("Quarters")
plt.xlabel("Employee")
plt.show()
```
